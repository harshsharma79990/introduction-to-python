1. What is a parameter?

Answer1- A parameter is a value used by a machine learning model to make predictions. These values
are learned automatically from the training data during the learning process. Parameters help the model to understand the relationship between input and output variables. For example, in a linear regression model, the weights and intercept are parameters that determine the prediction made by the model.

2. What is correlation? What does negative correlation mean?

Answer2- Correlation is a statistical measure that shows the strength and direction of the relationship between two variables. It helps us understand whether changes in one variable are associated with changes in another variable.
Negative correlation means that the two variables move in opposite directions. When one variable
increases, the other decreases. For example, as the amount of time spent exercising increases, body fat percentage may decrease. This indicates a negative correlation between exercise and body fat.

3. Define Machine Learning. What are the main components in Machine Learning?

Answer3- Machine Learning is a branch of Artificial Intelligence that enables computers to learn from data and improve their performance without being explicitly programmed for every task. Instead of following fixed rules, machine learning systems identify patterns in data and use them to make predictions or decisions.
The main components of Machine Learning are:
- Data- The information used for training the model.
- Features – The input variables used to make predictions.
- Model – The algorithm that learns patterns from data.
- Loss Function – Measures the prediction error.
- Optimizer – Reduces the error by adjusting parameters.
- Training Process – The process of teaching the model.
- Evaluation Metrics – Used to measure model performance.

4. How does loss value help in determining whether the model is good or not?

Answer4- Loss value is a measure of how far the model's predictions are from the actual values. It
helps us understand the accuracy of a model during training.
A high loss value indicates that the model is making large errors, while a low loss value indicates that
the model's predictions are close to the actual results. During training, the goal of the model is to minimize the loss value. Therefore, a lower loss generally indicates a better-performing model.

15. What is causation? Explain the difference between correlation and causation with an example.

Answer15- Causation means that one event directly causes another event to occur. It represents a true cause-and-effect relationship.
Correlation only shows that two variables are related, but it does not prove that one causes the other.
Example:
Ice cream sales and drowning incidents both increase during summer.
These two variables are correlated because they increase together. However, buying ice cream does not cause drowning. The actual cause is hot weather, which increases both ice cream consumption and swimming activities.
Therefore, correlation does not always imply causation.

16. What is an Optimizer? What are different types of optimizers?

Answer16- An optimizer is an algorithm used to improve a machine learning model by reducing the
loss function. It adjusts the model parameters so that prediction errors become smaller.
Types of Optimizers
- Gradient Descent
Uses the entire dataset to update parameters.
- Stochastic Gradient Descent (SGD)
Updates parameters after processing each training example.
- Mini-Batch Gradient Descent
Uses small batches of data for parameter updates.
- Adam Optimizer
Combines the benefits of momentum and adaptive learning rates. It is widely used because it is fast and efficient. Optimizers play a key role in helping models learn effectively.

21. What is Feature Scaling? How does it help in Machine Learning?

Answer21- Feature Scaling is the process of transforming numerical data so that all features have a similar range of values. In many datasets, some features may have very large values while others have very small values. This difference can affect the performance of machine learning algorithms.
For example, consider a dataset containing:
• Age: 18 to 60
• Salary: 20,000 to 1,00,000
Here, salary values are much larger than age values. As a result, the model may give more importance to salary and ignore age. Feature scaling helps solve this problem.
Benefits of Feature Scaling:

- Improves the performance of machine learning algorithms.

- Speeds up the training process.

- Helps the model converge faster.

- Improves prediction accuracy in many algorithms.
Feature scaling is especially important in algorithms such as K-Nearest Neighbors (KNN), Support Vector Machines (SVM) and Neural Networks.
Therefore, feature scaling helps create a balanced dataset and improves the efficiency of machine
learning models.

25. Explain Data Encoding.

Answer25- Data Encoding is the process of converting categorical or textual data into numerical values
so that machine learning algorithms can understand and process it.
Most machine learning models work only with numbers. Therefore, categories such as names, colors,
cities, or genders must be converted into numerical form before model training.
Example
Gender Encoded Value
Male   0
Female 1
In this example, the categorical values are converted into numbers.
Common Encoding Techniques
1. Label Encoding

Each category is assigned a unique numerical value.
Example:
• Red = 0
• Blue = 1
• Green = 2

2. One-Hot Encoding

Creates separate columns for each category and uses 0 or 1 to represent the presence of a category.

3. Ordinal Encoding
Used when categories have a natural order.
Example:
• Low = 1
• Medium = 2
• High = 3

Importance of Data Encoding
-  Makes categorical data understandable to machine learning models.
-  Improves model performance.
-  Enables proper data processing and analysis.
-  Helps algorithms work efficiently with non-numerical data.

Therefore, data encoding is an essential preprocessing step in machine learning because it converts categorical information into a numerical format suitable for model training and prediction.
