Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

Answer:

Deep Learning is a branch of Artificial Intelligence (AI) and Machine Learning that uses neural networks with many layers to learn patterns from data. It helps computers perform tasks such as image recognition, speech recognition, language translation, and prediction.

The idea of Deep Learning started from artificial neural networks, which were inspired by the working of the human brain. Earlier, computers had limited processing power and data, so deep neural networks were not widely used. With the availability of large datasets, powerful computers, and Graphics Processing Units (GPUs), Deep Learning became more practical and achieved excellent results in many fields.

The main difference between Deep Learning and traditional Machine Learning is that traditional Machine Learning usually requires humans to select and create important features from data. Deep Learning can automatically learn these features by itself from large amounts of data. Deep Learning generally performs better on complex tasks but requires more data and computational power.

Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

Answer:

A Perceptron is the simplest type of artificial neural network and is considered the basic building block of Deep Learning. It is mainly used for binary classification problems where the output belongs to one of two classes.

A Perceptron consists of:

- Input Layer – Receives input values.
- Weights – Each input is assigned a weight that represents its importance.
- Summation Function – Multiplies inputs by their weights and adds them together.
- Activation Function – Decides the final output based on the calculated value.
- Output – Produces the final prediction.

Working of a Perceptron:

Inputs are provided to the perceptron.
Each input is multiplied by its corresponding weight.
The weighted values are added together.
An activation function is applied.
The final output is generated.

Limitations of Perceptron:

It can only solve linearly separable problems.
It cannot solve complex problems like the XOR problem.
It has only one layer, so its learning capability is limited.
It cannot capture complicated patterns in data.

Because of these limitations, multi-layer neural networks were developed.

Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

Answer:

An activation function is a mathematical function used in neural networks to determine whether a neuron should be activated or not. It introduces non-linearity into the network, allowing the model to learn complex patterns and relationships in data.

Without activation functions, a neural network would behave like a simple linear model and would not be able to solve complex problems.

- Sigmoid Function

The Sigmoid function converts input values into outputs between 0 and 1.

Advantages:

Easy to understand.
Useful for binary classification problems.
Output can be interpreted as probability.

Disadvantages:

Training can become slow because of the vanishing gradient problem.
Not suitable for deep neural networks.

- ReLU (Rectified Linear Unit)

The ReLU function returns the input value if it is positive; otherwise, it returns zero.

Advantages:

Simple and computationally efficient.
Trains deep networks faster.
Helps reduce the vanishing gradient problem.

Disadvantages:

Some neurons may stop learning if they always produce zero output.

- Tanh Function

The Tanh function produces output values between -1 and 1.

Advantages:

Output is centered around zero.
Often performs better than Sigmoid.

Disadvantages:

Can also suffer from the vanishing gradient problem.

- Comparison:

Sigmoid: Output range 0 to 1, mainly used for probabilities.

ReLU: Output range 0 to positive values, commonly used in hidden layers.

Tanh: Output range -1 to 1, provides better learning than Sigmoid in many cases.

Question 4: What is the difference between Loss Function and Cost Function in neural networks? Provide examples.

Answer:

Loss Function and Cost Function are used to measure how well a neural network is performing by comparing predicted values with actual values.

- Loss Function

A Loss Function calculates the error for a single training example. It shows how far the prediction is from the actual value.

Example:

Suppose the actual value is 10 and the predicted value is 8. The loss function measures the difference between them.

- Cost Function

A Cost Function is the average of the loss values for all training examples in the dataset. It provides an overall measure of model performance.

Example:

If a dataset contains 100 samples, the cost function calculates the average error across all 100 samples.

- Difference Between Loss and Cost Function:

Loss Function measures error for a single data point.

Cost Function measures average error for the entire dataset.

Loss helps evaluate individual predictions.

Cost helps evaluate the overall model performance.

- Examples:

Mean Squared Error (MSE)

Binary Cross-Entropy

Categorical Cross-Entropy


These functions help the neural network learn and improve its predictions during training.

Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

Answer:

Optimizers are algorithms used to improve the performance of a neural network by adjusting its weights and biases during training. Their main goal is to reduce the error or loss function and help the model achieve better accuracy.

Without an optimizer, a neural network would not know how to update its parameters and learn from data.

1. Gradient Descent

Gradient Descent updates weights in the direction that reduces the loss.

Advantages:

Simple and easy to understand.
Forms the basis of many optimization techniques.

Disadvantages:

Can be slow on large datasets.
May get stuck in local minima.

2. RMSprop

RMSprop adjusts the learning rate automatically for each parameter during training.

Advantages:

Faster convergence than basic Gradient Descent.
Works well for complex neural networks.

Disadvantages:

Requires tuning of parameters.

3. Adam (Adaptive Moment Estimation)

Adam combines the advantages of Gradient Descent and RMSprop. It adapts the learning rate and uses information from previous updates.

Advantages:

Fast and efficient.
Works well for most Deep Learning tasks.
Requires less manual tuning.

Disadvantages:

Slightly more complex than basic Gradient Descent.
Comparison
Gradient Descent: Simple but slower.
RMSprop: Faster and adapts learning rates.
Adam: Most popular optimizer, combines speed and efficiency, and usually gives better results.

```python
'''Question 6: Write a Python program to implement a single-layer perceptron
from scratch using NumPy to solve the logical AND gate.'''
import numpy as np

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

weights = np.array([1, 1])
bias = -1.5

def perceptron(x):
    if x >= 0:
        return 1
    else:
        return 0

for i in inputs:
    total = np.dot(i, weights) + bias
    output = perceptron(total)
    print(i, "->", output)
```

```python
'''Question 7: Implement and visualize Sigmoid, ReLU, Tanh activation functions
using Matplotlib.
(Include your Python code and output in the code box below.)
'''
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

# Activation Functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

# Plot graphs
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")

plt.title("Activation Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
'''Question 8: Use Keras to build and train a simple multilayer neural network
 on the MNIST digits dataset. Print the training accuracy.
(Include your Python code and output in the code box below.)'''
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize data
X_train = X_train / 255.0

# Create model
model = Sequential()

model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(X_train, y_train, epochs=3)

# Print accuracy
print("Training Accuracy:", history.history['accuracy'][-1])
```

```python
'''Question 9: Visualize the loss and accuracy curves for a neural network model
 trained on the Fashion MNIST dataset. Interpret the training behavior.'''
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize data
X_train = X_train / 255.0

# Create model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(X_train, y_train, epochs=5)

# Plot Accuracy
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()

# Plot Loss
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

Question 10: You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.

Explain your real-time data science workflow:

● How would you design a deep learning model (perceptron or multilayer NN)?

● Which activation function and loss function would you use, and why?

● How would you train and evaluate the model, considering class imbalance?

● Which optimizer would be suitable, and how would you prevent overfitting?

Answer10: For detecting fraudulent bank transactions, I would use a Multilayer Neural Network (MLP) because the dataset is large and contains many features such as transaction amount, merchant ID, and customer location. A multilayer neural network can learn complex patterns better than a single perceptron.

- Designing the Deep Learning Model

The model would contain:

Input Layer to receive transaction data
One or more Hidden Layers to learn patterns
Output Layer to predict whether the transaction is fraudulent or legitimate

The output layer would have one neuron because there are only two classes: Fraud and Legitimate.

- Activation Function and Loss Function

Activation Function: ReLU

I would use ReLU in the hidden layers because it is simple, fast, and helps the network learn efficiently.

Loss Function: Binary Cross-Entropy

Since the problem has only two classes (fraud or legitimate), Binary Cross-Entropy is the most suitable loss function. It measures how far the predictions are from the actual values.

- Training and Evaluation

The dataset is imbalanced because fraudulent transactions are usually much fewer than legitimate transactions.

- To handle this problem, I would:

Use class weights so fraud cases get more importance during training.
Split the data into training and testing datasets.
Evaluate the model using Accuracy, Precision, Recall, and F1-Score.

Recall is especially important because missing a fraud transaction can cause financial loss.

- Optimizer and Overfitting Prevention

Optimizer: Adam

Adam is a good choice because it trains faster and usually gives good results without much tuning.

- For Preventing Overfitting:

Use Dropout layers.

Use Early Stopping to stop training.

Train with sufficient data.

Validate the model on unseen data.
