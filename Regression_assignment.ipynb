Question 1: What is Simple Linear Regression?

Answer:

Simple Linear Regression is a statistical and machine learning technique used to find the relationship between one independent variable (input) and one dependent variable (output). It helps us predict the value of one variable based on another.

It represents the relationship using a straight line called the regression line. The equation of this line is:

Y = a + bX

(where a is the intercept and b is the slope of the line.)

Question 2: What are the key assumptions of Simple Linear Regression?

Answer: Simple Linear Regression works best when these assumptions are reasonably true:

- Linearity: The relationship between X and Y should be approximately a straight line.

- Independence: The observations should be independent of each other.

- Constant variance: The spread of errors should remain roughly the same for all values of X.

- Normality of errors: The prediction errors (residuals) should be approximately normally distributed.

- No extreme outliers: Very unusual data points can strongly affect the regression line and reduce accuracy.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer:

Heteroscedasticity occurs when the spread or variance of errors is not constant across all values of the independent variable. In simple terms, the prediction errors become larger or smaller as the value of the input variable changes.

It is important to address heteroscedasticity because:

- It can make the regression results less reliable.

- The estimated coefficients may become inefficient.

- Confidence intervals may give misleading results.

- Predictions may be less accurate.

To reduce heteroscedasticity, data transformations, feature engineering, or specialized regression techniques can be used.

Question 4: What is Multiple Linear Regression?

Answer: Multiple Linear Regression is an extension of simple linear regression that uses two or more input variables to predict one output variable.

For example, a student's exam score might depend on study hours, attendance, and previous marks. Multiple linear regression combines all these factors to make a prediction.

It helps us understand how each variable affects the outcome while keeping the other variables constant.

Question 5: What is Polynomial Regression, and how does it differ from Linear Regression?

Answer:

Polynomial Regression is a regression technique used when the relationship between the input and output variables is not a straight line. Instead, it fits a curved line to the data.

Difference between Linear Regression and Polynomial Regression:

- Linear Regression fits a straight line to the data.

- Polynomial Regression fits a curved line to the data.

- Linear Regression is suitable when variables have a linear relationship.

- Polynomial Regression is suitable when the relationship is nonlinear.

- Polynomial Regression can capture more complex patterns than Linear Regression.



```python
'''
Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:

● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points.

(Include your Python code and output in the code box below.)

Answer:
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print equation values
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Plot data points
plt.scatter(X, Y, label="Actual Data")

# Plot regression line
plt.plot(X, Y_pred, label="Regression Line")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()

```

```python
'''
Question 7: Fit a Multiple Linear Regression model on this sample data:

 ● Area = [1200, 1500, 1800, 2000]

 ● Rooms = [2, 3, 3, 4]

 ● Price = [250000, 300000, 320000, 370000]

 Check for multicollinearity using VIF and report the results.

 (Include your Python code and output in the code box below.)

  Answer:
  '''

import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Independent variables
X = data[['Area', 'Rooms']]
Y = data['Price']

# Train model
model = LinearRegression()
model.fit(X, Y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                   for i in range(X.shape[1])]

print(vif_data)
```

```python
'''
Question 8: Implement polynomial regression on the following data:

● X = [1, 2, 3, 4, 5] 3

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.

(Include your Python code and output in the code box below.)

Answer:
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Convert to polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X_poly, Y)

# Predictions
Y_pred = model.predict(X_poly)

# Plot
plt.scatter(X, Y, label="Actual Data")
plt.plot(X, Y_pred, label="Polynomial Curve")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

```python
'''
Question 9: Create a residuals plot for a regression model trained on this data:
 ● X = [10, 20, 30, 40, 50]
 ● Y = [15, 35, 40, 50, 65]
 Assess heteroscedasticity by examining the spread of residuals.
(Include your Python code and output in the code box below.)
Answer:
'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Residuals
residuals = Y - Y_pred

# Residual plot
plt.scatter(Y_pred, residuals)
plt.axhline(y=0)

plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()

print("Residuals:", residuals)
```
