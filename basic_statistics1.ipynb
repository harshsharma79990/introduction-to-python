Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples?

Answer: Statistics is mainly divided into two parts: descriptive statistics and inferential statistics. Both are used to work with data, but their purpose is different.
1. Descriptive Statistics
Descriptive statistics is used to describe, summarize, and present data in a simple and understandable form.
It only talks about the data that is already collected and does not make predictions about the future.
Main purpose:
To show what the data looks like.
Common tools used:

•	Mean (average)

•	Median

•	Mode

•	Range

•	Tables, graphs, bar charts, pie charts

Example:
Suppose the marks of 5 students are:
40, 50, 60, 70, 80
•	The average (mean) marks = 60
•	A bar graph is drawn to show marks of each student

Here, we are only describing the given marks, not predicting anything.
So, this is descriptive statistics.

2. Inferential Statistics

Inferential statistics is used to draw conclusions or make predictions about a large group (population) using a small group (sample).
Main purpose:
To guess or predict what may happen for the whole population based on sample data.

Common tools used:

•	Probability

•	Hypothesis testing

•	Confidence interval

•	Regression analysis

Example:
Suppose a teacher checks the marks of 30 students out of 300 students in a college and finds that most students scored above 60%.
Using this sample result, the teacher concludes that
most students in the entire college performed well.

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer: Sampling is the process of selecting a small group (sample) from a large group (population) to study and analyze.
Studying the entire population is often difficult, time-consuming, and costly, so instead we study a sample and use the results to understand the whole population.

Example:
If a college has 1,000 students, it is hard to ask every student about their study habits.
So, we select 100 students as a sample.
This process is called sampling.


Types of Sampling
Two common types of sampling are:

Random Sampling

Stratified Sampling

1. Random Sampling

In random sampling, every member of the population has an equal chance of being selected.
Selection is done purely by chance, without any bias.
Key features:

•	Each person has equal opportunity

•	Simple and easy method

•	No favoritism or bias

Example:

If there are 200 students in a class and 20 students are selected by:

•	Lottery method

•	Random number generator

Then this is random sampling.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important?

Answer: Mean, median, and mode are called measures of central tendency. They are used in statistics to find a single value that represents the center or typical value of a given set of data. These measures help us understand and summarize large amounts of data in a simple way.
- Mean

The mean is the average of all the values in a data set. It is calculated by adding all the values and then dividing the total by the number of values.
For example, if the marks of five students are 10, 20, 30, 40, and 50, the mean is found by adding all the marks and dividing by 5. The mean helps us understand the overall performance of the group.
- Median

The median is the middle value of a data set when the values are arranged in ascending or descending order. If the number of values is even, the median is the average of the two middle values.
For example, if the data is 5, 10, 15, 20, and 25, the median is 15 because it lies in the middle. The median is very useful when the data has very high or very low values, as it is not affected by extreme values.
- Mode

The mode is the value that occurs most frequently in a data set. A data set can have one mode, more than one mode, or no mode at all.
For example, in the data 2, 4, 4, 6, 8, the mode is 4 because it appears more times than the other numbers. The mode is helpful when we want to know the most common or popular value.
- Importance of mean, median, and mode

These measures are important because they help in understanding data quickly and easily. Instead of looking at many numbers, we can use one value to represent the whole data set. The mean shows the overall average, the median shows the central position, and the mode shows the most common value. They are widely used in daily life, such as calculating average marks, salaries, income, or finding the most common choice in a survey. Using these measures makes data analysis simple, clear, and meaningful.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer: Skewness and kurtosis are statistical measures that describe the shape of a data distribution. They help us understand how data values are spread and how the distribution looks beyond just the average.
Skewness shows whether the data is symmetrical or tilted to one side. In a perfectly symmetrical distribution, the left and right sides are equal. When data is not balanced, it is said to be skewed.
- There are three cases of skewness:

If skewness is zero, the data is symmetrical.

If skewness is positive, the data is positively skewed.

If skewness is negative, the data is negatively skewed.

In a positively skewed distribution, most of the data values are concentrated on the left side, while a few very large values stretch the graph to the right side. This creates a long tail on the right. In such cases, the mean is usually greater than the median.

Example: Income data is often positively skewed because most people earn average or low income, while a few people earn very high income.
- Kurtosis shows how peaked or flat a distribution is compared to a normal distribution. It tells us about the presence of extreme values (very high or very low values) in the data.



```python
# Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers?
# Answer:
# Python program to calculate mean, median, and mode

numbers = [10, 20, 20, 30, 40, 50]

# Mean
mean = sum(numbers) / len(numbers)

# Median
numbers.sort()
n = len(numbers)

if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
    median = numbers[n//2]

# Mode
mode = max(set(numbers), key=numbers.count)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
''' Question 6: Compute the covariance and correlation coefficient between
the following two datasets provided as lists in Python:
 list_x = [10, 20, 30, 40, 50]
 list_y = [15, 25, 35, 45, 60]
 '''
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance
covariance = np.cov(list_x, list_y)[0][1]

# Correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```
