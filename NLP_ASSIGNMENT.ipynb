Question 1: What is Computational Linguistics and how does it relate to NLP?

Answer:

Computational Linguistics is a field that combines computer science and linguistics (the study of language). It focuses on teaching computers how human language works so that they can understand, process, and generate text or speech.

Natural Language Processing (NLP) is a practical application of Computational Linguistics. Computational Linguistics provides the language rules, theories, and knowledge, while NLP uses these concepts to build real-world applications such as chatbots, language translators, voice assistants, and text analysis systems.

In simple words, Computational Linguistics helps us understand language scientifically, and NLP uses that understanding to make computers work with human language.

Question 2: Briefly describe the historical evolution of Natural Language Processing.

Answer:

The development of NLP has gone through several stages over the years.

In the 1950s, researchers started working on machine translation and basic language processing systems. Computers followed simple hand-written grammar rules to process language.

During the 1970s and 1980s, rule-based systems became popular. These systems used a large number of language rules created by experts. However, they were difficult to maintain and could not handle all language variations.

In the 1990s, statistical methods were introduced. Computers started learning patterns from large collections of text instead of relying only on rules. This improved the accuracy of NLP systems.

In the 2000s and 2010s, machine learning techniques became widely used. Computers learned from huge datasets and performed tasks such as translation, sentiment analysis, and speech recognition more effectively.

Today, NLP uses advanced Artificial Intelligence and Deep Learning models. Modern systems like virtual assistants, chatbots, and language generation tools can understand and generate human-like text with high accuracy.

Question 4: What is text normalization and why is it essential in text processing tasks?

Answer:

Text normalization is the process of converting text into a standard and consistent format before it is analyzed by a computer.

This process may include converting all letters to lowercase, removing extra spaces, correcting spelling variations, removing punctuation, and expanding abbreviations.

For example:

Original Text: "HELLO!!! How R U?"

Normalized Text: "hello how are you"

Text normalization is important because text data often comes in different formats. If the text is not standardized, computers may treat similar words as different words. Normalization improves data quality, reduces errors, and helps NLP models produce more accurate results.

Question 5: Compare and contrast stemming and lemmatization with suitable examples.

Answer:

Stemming and lemmatization are two techniques used in Natural Language Processing (NLP) to reduce words to their basic form. This helps computers treat similar words as the same word during text analysis.

Stemming removes the ending part of a word to get a shorter form. It does not check whether the resulting word is a proper dictionary word or not.

- Examples of Stemming:

playing becomes play

played becomes play

studying becomes studi

Because stemming simply cuts off word endings, the final word may sometimes look incorrect.

Lemmatization reduces a word to its actual dictionary form by considering its meaning and grammar. Therefore, the result is usually a proper and meaningful word.

- Examples of Lemmatization:


playing becomes play

studies becomes study

better becomes good

The main difference is that stemming is a simple and faster method, while lemmatization is more accurate because it understands the meaning of words. Stemming may produce incomplete words, whereas lemmatization generally produces correct dictionary words.


```python
'''Question 6: Write a Python program that uses regular expressions (regex) to extract all
email addresses from the following block of text:
“Hello team, please contact us at support@xyz.com for technical issues, or reach out to
our HR at hr@xyz.com. You can also connect with John at john.doe@xyz.org and jenny
via jenny_clarke126@mail.co.us. For partnership inquiries, email partners@xyz.biz.”'''
import re

text = """
Hello team, please contact us at support@xyz.com for technical issues,
or reach out to our HR at hr@xyz.com.
You can also connect with John at john.doe@xyz.org and jenny via
jenny_clarke126@mail.co.us.
For partnership inquiries, email partners@xyz.biz.
"""

emails = re.findall(r'[\w\.-]+@[\w\.-]+\.\w+', text)

print("Email Addresses Found:")
for email in emails:
    print(email)
```

```python
'''Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:
“Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.”
'''
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')

text = """
Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.
"""

# Tokenization
tokens = word_tokenize(text)

print("Tokens:")
print(tokens)

# Frequency Distribution
freq = FreqDist(tokens)

print("\nTop 10 Most Frequent Tokens:")
print(freq.most_common(10))
```

```python
'''Question 8: Create a custom annotator using spaCy or NLTK that identifies and labels
proper nouns in a given text'''
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

text = "John lives in New York and works for Google."

# Tokenize the text
words = word_tokenize(text)

# Part-of-speech tagging
tagged_words = pos_tag(words)

print("Proper Nouns Found:")

for word, tag in tagged_words:
    if tag == "NNP":
        print(word, "- Proper Noun")
```
