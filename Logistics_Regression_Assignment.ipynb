Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

Answer:

Simple Linear Regression (SLR) is a statistical technique used to find the relationship between one independent variable (X) and one dependent variable (Y). It helps us understand how changes in one variable affect another variable.

The main purpose of Simple Linear Regression is to predict the value of the dependent variable based on the value of the independent variable. It also helps in identifying trends and making future predictions from existing data.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer:

Simple Linear Regression works properly when the following assumptions are satisfied:

- Linear Relationship – There should be a straight-line relationship between the independent variable and the dependent variable.
- Independence of Observations – The data observations should be independent of each other.
- Constant Variance (Homoscedasticity) – The errors should have a constant spread across all values of the independent variable.
- Normal Distribution of Errors – The prediction errors should be approximately normally distributed.
- No Significant Outliers – Extreme values should not strongly affect the regression line.

These assumptions help ensure that the model gives accurate and reliable predictions.

Question 5: What is the method of least squares in linear regression?

Answer:

The Method of Least Squares is a technique used to find the best-fitting regression line for a set of data points.

In a dataset, the actual values and predicted values are usually different. The difference between them is called an error.

The least squares method finds the regression line by minimizing the sum of the squares of these errors. Squaring the errors ensures that positive and negative errors do not cancel each other out and gives more importance to larger errors.

In simple words, this method chooses the line that passes as close as possible to all data points, resulting in more accurate predictions.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Answer:

Logistic Regression is a machine learning and statistical technique used for classification problems, where the goal is to predict a category rather than a numerical value. It is commonly used to determine outcomes such as whether a student will pass or fail, whether an email is spam or not spam, or whether a customer will purchase a product. Logistic Regression calculates the probability of an event occurring and gives an output between 0 and 1. Based on this probability, the result is classified into a specific category. In contrast, Linear Regression is used to predict continuous numerical values such as house prices, sales revenue, or exam marks. The main difference is that Linear Regression predicts a numeric value, while Logistic Regression predicts a category or class. Therefore, Linear Regression is used for prediction tasks, whereas Logistic Regression is used for classification tasks.

Question 7: Name and briefly describe three common evaluation metrics for regression models.

Answer:

- Mean Absolute Error (MAE) – It measures the average absolute difference between the actual values and the predicted values. A smaller MAE indicates better model performance.
- Mean Squared Error (MSE) – It calculates the average of the squared differences between actual and predicted values. Larger errors receive more penalty because they are squared.
- Root Mean Squared Error (RMSE) – It is the square root of MSE. RMSE gives the prediction error in the same units as the target variable, making it easier to understand.

Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer:

R-squared, also known as the coefficient of determination, measures how well a regression model explains the variation in the dependent variable. Its value ranges from 0 to 1.

An R-squared value close to 1 indicates that the model explains most of the variation in the data, while a value close to 0 means that the model does not explain the data well. Therefore, R-squared helps us understand how well the regression model fits the dataset.


```python
'''
Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

Answer:
'''
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, Y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Answer:

In a Simple Linear Regression model, the coefficients help us understand the relationship between the independent variable and the dependent variable. The slope coefficient shows how much the dependent variable changes when the independent variable increases by one unit. A positive coefficient means that the dependent variable increases as the independent variable increases, while a negative coefficient means that the dependent variable decreases. The intercept represents the value of the dependent variable when the independent variable is zero.

For example, in the equation:

Y = 10 + 3X

the intercept is 10, which means Y will be 10 when X is 0. The coefficient (slope) is 3, which means Y increases by 3 units whenever X increases by 1 unit. This helps us understand the impact of the independent variable on the predicted outcome.
