1. What is the difference between a function and a method in Python?

Answer: A function and a method in Python are both reusable blocks of code, but they are used in different contexts. A function is defined independently and can be called directly by using its name. It performs a specific task and does not belong to any particular object. A method is actually a function that is associated with an object and is called through that object. Methods usually work on the data contained inside the object and are part of object oriented programming. The key idea is that functions exist on their own, while methods are connected to objects and operate on them. For example, defining a function add(a, b) that returns a + b shows a function being called as add(2, 3). A method example would be using text.upper() where upper is a method applied to the string object text.

2. Explain the concept of function arguments and parameters in Python.

Answer: Function parameters and arguments are concepts that describe how information is passed into functions. Parameters are the variables listed when defining a function and they act as placeholders for data. Arguments are the actual values supplied when the function is called. When execution begins, the arguments are assigned to the parameters so that the function can use them. This mechanism allows one function to operate on different inputs without rewriting code. Understanding this distinction helps in designing flexible programs. For example, in the function definition def multiply(x, y): x and y are parameters, and when calling multiply(4, 5), the values 4 and 5 are arguments passed into the function.

3. What are the different ways to define and call a function in Python?

Answer: Python allows functions to be defined and called in several ways, giving programmers flexibility. The most common way is using the def keyword to define a function and then calling it by its name. Functions may include default parameter values so they still work if certain arguments are not provided. They can be called using positional arguments or keyword arguments where values are matched by name. Functions can also accept an unknown number of arguments using special syntax. Anonymous functions can be written using lambda for quick operations. These variations allow functions to be reused in many programming situations. 4. What is the purpose of the return statement in a Python function?

Answer: The return statement in Python functions is used to send a value back to the place from which the function was called. It ends the execution of the function and provides output that can be stored or reused elsewhere in the program. Without return, a function completes its task but gives back a default value of None, which may not be useful. Using return improves modularity because results can be passed between different parts of a program. It also allows functions to perform calculations and supply answers rather than just displaying output. For example, defining def square(n): return n * n allows calling result = square(6) and storing the value 36 in result.

5. What are iterators in Python and how do they differ from iterables?

Answer:Iterables and iterators are concepts that allow sequential access to data. An iterable is any object that contains multiple values and can be looped over, such as lists, tuples, or strings. An iterator is the mechanism that actually goes through each value step by step and remembers its current position. When a loop runs, Python internally creates an iterator from the iterable to fetch items one at a time. The iterable represents the collection, while the iterator controls traversal. Understanding this difference helps when working with custom data processing. For example, a list numbers = [1, 2, 3] is an iterable, and using iter(numbers) creates an iterator that can retrieve values using next() until the list is exhausted.

 6. Explain the concept of generators in Python and how they are defined.

 Answer: Generators in Python are special functions that produce values one at a time instead of returning all results at once. They are defined similarly to normal functions but use the yield keyword. When yield is encountered, the function pauses and saves its state, then resumes from that point when called again. This approach is memory efficient because values are created only when needed rather than stored in advance. Generators are particularly useful when dealing with large data streams or infinite sequences. They simplify iteration logic and reduce resource usage. For example, defining def count_up(n): for i in range(n): yield i creates a generator that produces numbers gradually when used in a loop.

  7. What are the advantages of using generators over regular functions?
  
  Answer: Generators offer several advantages compared to regular functions that return complete lists. They require less memory since they do not store all values at once. This makes them efficient when processing large datasets or continuous data streams. They also improve speed because values are generated only when requested. Another advantage is cleaner code, since state management is handled automatically by the generator mechanism. These benefits make them suitable for tasks involving iteration over large ranges or files. For example, generating squares of numbers using a generator expression like (x*x for x in range(1000000)) avoids storing a huge list in memory and calculates each value only when accessed.

  8. What is a lambda function in Python and when is it typically used?
  
  Answer: A lambda function in Python is a small unnamed function defined in a single expression. It is commonly used for short tasks where writing a full function definition would be unnecessary. Lambda functions are often passed as arguments to higher order functions or used in quick transformations.
```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even(numbers):
    total = 0
    for n in numbers:
        if n % 2 == 0:
            total = total + n
    return total

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even(nums))
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(text):
  reversed_text = ""
  for ch in text:
    reversed_text = ch + reversed_text
  return reversed_text

print(reverse_string("hello"))
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(values):
  result = []
  for v in values:
    result.append(v * v)
  return result

data = [1, 2, 3, 4]
print(square_list(data))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime(num):
    if num < 2 or num > 200:
        return "Number out of range"
    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"
    return "Prime"

print("Prime check result:", check_prime(37))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exp):
    i = 0
    while i <= exp:
        yield 2 ** i
        i += 1

for value in powers_of_two(5):
    print(value)
```

```python
'''8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.'''
data = [(1, 3), (2, 1), (4, 2), (3, 5)]
sorted_data = sorted(data, key=lambda x: x[1])
print("Sorted list:", sorted_data)
```

```python
# 9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Fahrenheit values:", fahrenheit)
```
