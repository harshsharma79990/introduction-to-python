Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:

Anomaly Detection is the process of finding unusual data points that are different from the rest of the data. These unusual observations are called anomalies or outliers. Anomaly detection is commonly used in fraud detection, network security and fault detection systems.

Types of Anomalies:

1. Point Anomaly
A single data point that is very different from other data points.

Example: If most students score between 60 and 90 marks, but one student scores 5 marks, then 5 is a point anomaly.

2. Contextual Anomaly
A data point that is abnormal in a specific context, such as time or location.

Example: A temperature of 15°C is normal in winter but unusual in summer. Therefore, it is a contextual anomaly in summer.

3. Collective Anomaly
A group of data points that together behave abnormally, even though individual points may appear normal.

Example: A sudden series of unusual credit card transactions occurring one after another may indicate fraud. Each transaction may look normal alone, but together they are suspicious.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer:

- Isolation Forest
Isolation Forest detects anomalies by randomly splitting data points. Since anomalies are different from normal data, they get isolated more quickly.

- Use Case: Detecting fraudulent bank transactions or unusual customer behavior.

- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
DBSCAN groups closely packed data points into clusters. Points that do not belong to any cluster are treated as anomalies.

- Use Case: Finding unusual locations in GPS data or detecting noise in datasets.

- Local Outlier Factor (LOF)
LOF compares the density of a data point with the density of its neighboring points. If a point has much lower density than its neighbors, it is considered an anomaly.

- Use Case: Detecting unusual customer purchases or abnormal network activity.

In simple words,

Isolation Forest isolates anomalies

DBSCAN finds anomalies outside clusters

LOF finds anomalies based on neighborhood density.

Question 3: What are the key components of a Time Series? Explain each with one example.

Answer:

A Time Series is a sequence of data collected over time, such as daily sales or monthly temperature records.

The main components of a time series are:

1. Trend

Trend represents the long-term increase or decrease in data.

Example: A company's sales increasing every year.

2. Seasonality

Seasonality refers to patterns that repeat at regular intervals.

Example: Ice cream sales increase every summer and decrease every winter.

3. Cyclical Component

Cyclical changes occur over longer periods and are usually influenced by economic or business conditions.

Example: An increase and decrease in car sales during economic booms and recessions.

4. Residual (Random Variation)

Residuals are irregular fluctuations that cannot be explained by trend or seasonality.

Example: A sudden drop in sales due to a natural disaster.

Question 4: Define Stationary in Time Series. How can you test and transform a non-stationary series into a stationary one?

Answer:

A time series is called stationary when its statistical properties, such as mean and variance, remain constant over time.

For example, if monthly sales remain around the same average value year after year, the series is considered stationary.

Testing Stationarity

A common method is the Augmented Dickey-Fuller (ADF) Test.

If the p-value is less than 0.05, the series is stationary.
If the p-value is greater than 0.05, the series is non-stationary.
Transforming a Non-Stationary Series

- Some common methods are:

Differencing – Subtract the previous value from the current value.

Log Transformation – Reduce large variations in data.

Removing Trend – Eliminate long-term growth or decline.

Removing Seasonality – Remove repeating seasonal patterns.

These techniques help make the data stationary so that forecasting models can work better.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer:

- AR (AutoRegressive)

AR uses past values of the series to predict future values.

Example: Predicting today's sales using sales from previous days.

- MA (Moving Average)

MA uses past prediction errors to forecast future values.

Example: Forecasting stock prices using previous forecasting errors.

- ARIMA (AutoRegressive Integrated Moving Average)

ARIMA combines AR and MA and also includes differencing to make data stationary.

Example: Forecasting monthly sales with trend but without seasonality.

- SARIMA (Seasonal ARIMA)

SARIMA extends ARIMA by handling seasonal patterns.

Example: Predicting monthly electricity consumption that follows yearly seasonal patterns.

- SARIMAX (Seasonal ARIMA with Exogenous Variables)

SARIMAX is similar to SARIMA but also includes external variables that affect predictions.

Example: Predicting product sales using past sales data and advertising expenditure.






```python
'''
Question 6: Load a time series dataset (e.g., AirPassengers), plot the original
 series, and decompose it into trend, seasonality, and residual components.

Answer:
'''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
)

data['Month'] = pd.to_datetime(data['Month'])

data.set_index('Month', inplace=True)

plt.figure(figsize=(8, 4))
plt.plot(data['Passengers'])
plt.title("AirPassengers Dataset")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

result = seasonal_decompose(data['Passengers'],
                            model='multiplicative',
                            period=12)

result.plot()
plt.show()
```

```python
'''
Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

Answer:
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

np.random.seed(42)

normal_data = np.random.normal(50, 10, 100)

anomalies = [120, 130, 140, 10, 5]

data = np.concatenate([normal_data, anomalies])

df = pd.DataFrame({'Value': data})

model = IsolationForest(contamination=0.05, random_state=42)
df['Anomaly'] = model.fit_predict(df[['Value']])

plt.figure(figsize=(8,5))
plt.scatter(df.index, df['Value'], c=df['Anomaly'])
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Data Point")
plt.ylabel("Value")
plt.show()
```

```python
'''
Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

Answer:
'''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

data = pd.read_csv(url)

data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

model = SARIMAX(
    data['Passengers'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

results = model.fit()

forecast = results.forecast(steps=12)

plt.figure(figsize=(10,5))
plt.plot(data['Passengers'], label='Original Data')
plt.plot(forecast, label='12-Month Forecast')
plt.legend()
plt.title("SARIMA Forecast")
plt.show()
```

```python
'''
Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

Answer:
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

np.random.seed(42)

normal_data = np.random.normal(50, 10, 100)

outliers = [120, 130, 5, 10]

data = np.concatenate([normal_data, outliers])

df = pd.DataFrame({'Value': data})

lof = LocalOutlierFactor(n_neighbors=20)

df['Anomaly'] = lof.fit_predict(df[['Value']])

plt.figure(figsize=(8,5))
plt.scatter(df.index, df['Value'], c=df['Anomaly'])
plt.title("LOF Anomaly Detection")
plt.xlabel("Data Point")
plt.ylabel("Value")
plt.show()
```
